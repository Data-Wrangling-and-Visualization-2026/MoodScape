### PREPROCESSING

libs for preprocessing

```python
import numpy as np
import pandas as pd
import json
```

Tracks may lack lyrics inherently (e.g., instrumentals), not due to retrieval issues. These tracks will be included in the analysis.

```python
df = pd.read_csv('df_tracks.csv')
df.head()
```

```python
df.tail()
```

```python
df = df.drop(['source', 'track_id-2', 'created_at', 'updated_at', 'created_at-2', 'updated_at-2', 'storage_uri'], axis = 1)
df.head()
```

```python
df['dsp'][0]
```

```python
df_analysis = pd.json_normalize(df['dsp'].apply(json.loads))
df_analysis = df_analysis.drop(['key', 'duration_ms'], axis = 1)
df = pd.concat([df.drop('dsp', axis=1), df_analysis], axis=1)
df.head()
```

```python
df.to_csv('processed_tracks.csv', index=False)
```
